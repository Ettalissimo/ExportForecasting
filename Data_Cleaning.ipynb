```python
# Changing features names
years = np.arange(2000, 2024)
columns_value_poids = []
for year in range(2000, 2024):
    columns_value_poids.append(f'poids_{year}')
    columns_value_poids.append(f'valeur_{year}')
data_export.columns = ['Chapitre', 'produit','Continent','pays','Libellé du flux'] + columns_value_poids +['unamed']
```

```python
# Changing the order of the features
new_columns_order = ['Chapitre', 'produit','Continent','pays','Libellé du flux']+ [f'valeur_{year}' for year in years] + [f'poids_{year}' for year in years] + ['unamed']
data_export = data_export[new_columns_order]
```

##Data Processing

```python
# Removing the unnecessary features

# List of columns to remove
columns_to_remove = ['Chapitre', 'Libellé du flux', 'unamed','unamed']

# Drop the specified columns
data_export = data_export.drop(columns=columns_to_remove, errors='ignore')
```

```python
data_export.dtypes
```

Converting some features from object type to float

```python

# Select columns to convert to float (excluding the first three columns)
columns_to_convert = data_export.columns[3:][data_export.iloc[:, 3:].dtypes == 'object']

# Remove commas from selected columns
data_export[columns_to_convert] = data_export[columns_to_convert].replace(',', '.', regex=True)

# Convert selected columns to float
data_export[columns_to_convert] = data_export[columns_to_convert].astype(float)

print(data_export.dtypes)
```

On s'interesse par la suite aux 3 produits suivant: Tomate, Citron et melon

```python
tomate_values = data_export[data_export['produit'].str.contains('TOMATE', case=False)]
citron_values = data_export[data_export['produit'].str.contains('CITRON', case=False)]
melon_values = data_export[data_export['produit'].str.contains('MELON', case=False)]
```

```python
# here we get only the data frame with tomato, citron and melon products
filtered_data = data_export[data_export['produit'].isin(tomate_values['produit']) |
                            data_export['produit'].isin(citron_values['produit']) |
                            data_export['produit'].isin(melon_values['produit'])]

filtered_data
```

```python
# Update product names containing 'tomato' to 'tomato'
filtered_data.loc[filtered_data['produit'].str.contains('TOMATE', case=False), 'produit'] = 'TOMATE'
filtered_data.loc[filtered_data['produit'].str.contains('CITRON', case=False), 'produit'] = 'CITRON'
filtered_data.loc[filtered_data['produit'].str.contains('MELON', case=False), 'produit'] = 'MELON'
```

```python
columns_pays = ['pays']

# Group by 'produit' and sum the values
filtred_data_continent = filtered_data.drop(columns=columns_pays, errors='ignore')
```

```python
import pandas as pd

# Assuming filtred_data_continent is your DataFrame

# Filter rows where the product is 'CITRON'
filtered_citron = filtred_data_continent[filtred_data_continent['produit'] == 'CITRON']

# Group by 'produit' and 'Continent', and sum up the numerical columns
data_citron = filtered_citron.groupby(['produit', 'Continent']).sum().reset_index()

# Print the resulting DataFrame
print(data_citron)
```

```python
# Create all possible combinations of products and continents
products = ['MELON', 'CITRON', 'TOMATE']

# Iterate over each product and continent combination
filtered_citron = filtred_data_continent[filtred_data_continent['produit'] == 'CITRON']
data_citron = filtered_citron.groupby(['produit', 'Continent']).sum().reset_index()

filtered_tomate = filtred_data_continent[filtred_data_continent['produit']== 'TOMATE']
data_tomate = filtered_tomate.groupby(['produit', 'Continent']).sum().reset_index()

filtered_melon = filtred_data_continent[filtred_data_continent['produit']== 'MELON'].groupby(['produit','Continent']).sum()
data_melon = filtered_melon.groupby(['produit', 'Continent']).sum().reset_index()


Data_continent = pd.concat([data_citron, data_tomate, data_melon], ignore_index= False)


Data_continent
```

```python
# Update product names containing 'tomato' to 'tomato'
filtered_data.loc[filtered_data['produit'].str.contains('TOMATE', case=False), 'produit'] = 'TOMATE'
filtered_data.loc[filtered_data['produit'].str.contains('CITRON', case=False), 'produit'] = 'CITRON'
filtered_data.loc[filtered_data['produit'].str.contains('MELON', case=False), 'produit'] = 'MELON'
```

```python
columns_continent = ['Continent']

# Group by 'produit' and sum the values
filtred_data_pays = filtered_data.drop(columns=columns_continent, errors='ignore')
filtred_data_pays = filtred_data_pays.groupby(['produit', 'pays']).sum()

# If you want to keep 'produit' as a column, not as an index
filtred_data_pays.reset_index(inplace=True)



filtred_data_pays
```

```python
# for now we are not interessted in region where we export
columns_region = ['Continent', 'pays']

filtred_data_without_region = filtered_data.drop(columns=columns_region, errors='ignore')

# Data without continent and countries coloumns
filtred_data_without_region
```
